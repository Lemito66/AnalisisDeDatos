## 3. Variables y tipos de datos
### Tipos de datos
Una variable es un espacio de memoria donde se almacenará el valor que le asignemos a dicha variable, podemos ver gráficamente en la imagen de abajo. Donde el **valor** de **20** se almacenará en el primer cuadrante de **memoria** y se refirirá a este cuadrante como **result** según este ejemplo. <br>
Se llaman **variables**, porque a pesar de reservar un espacio de memoria, el **valor** que se almacena en ese **espacio de memoria** puede cambiar *(variar)* con respecto al tiempo.
![variables](./images/variables.png)
Para crear una variable en Python se escribe el **nombre de la variable**, a continuación el signo igual **=** y luego el **valor** que se le quiere asignar a dicha variable. <br>
*El signo igual **=** es considerado como el **operador** de asignación.*

```python
result = 20
```

**TIP:** Las variables creadas en una **celda** de un **cuaderno** pueden ser utilizadas en **todas las celdas** de ese mismo cuadeno. <br>
**TIP:** Para mostrar en pantalla el **contenido** de una **variable** podemos usar la función **print()** y dentro el **nombre de la variable** que queremos imprimir.

```python
# result = 20
# print(result)
```

```python
print(result)
```

```python
print("result")
```

```python
print("result es ")
print(result)
```

```python
print("result es ")
print(result)
```

**TIP:** Para mostrar en pantalla el **contenido** de una **variable** usando celdas podemos simplemente poner el **nombre de la variable** que queremos imprimir en la última línea de la celda y esta se imprimirá.

```python
# cambiamos el contenido de result de 20 a 30.
result = 30
result
```

**TIP:** Los nombres de las variables tienen una propiedad llama **Case sensitive** donde el cambio entre minúsculas y mayúsculas afectan a las **variables**.

```python
print(Result)
```

### Tipos de datos
Los **valores** asignados a una **variable** son también conocidos como **datos**. <br>
Estos **datos** podrían ser de distintos **tipos**. Los **tipos de datos primitivos** son:
- **int** ejemplo: 2

```python
dos_int = 2
print(dos_int)
```

- **float** ejemplo: 2.0

```python
dos_float = 2.0
print(dos_float)
```

- **str** ejemplo: '2' o 'dos' o "2" o "dos"

```python
dos_str = '2'
print(dos_str)
```

```python
dos_str = "2"
print(dos_str)
```

```python
dos_str = 'dos'
print(dos_str)
```

```python
dos_str = "dos"
print(dos_str)
```

- **bool** ejemplo: True o False <br>

**TIP:** Los **valores** booleanos tienen la primera letra en mayúsculas.

```python
verdadero = True
print(verdadero)
```

```python
falso = False
print(falso)
```

```python
verdadero = true
```

```python
falso = false
```

Los **tipos de datos primitivos** son considerados de esta manera porque el resto de **tipos de datos** son construidos como una composición de estos **tipos de datos**. <br>

---

Hay otros **tipos de datos** no primitivos ampliamente utilizados en Python *(algunos de estos los veremos más adelante en el curso)*:
- **list** ejemplo: [1,2,3]
- **dict** ejemplo: { 'key':1,'key2':'value'}
- **Series**
- **DataFrame**
- **etc**

Hemos comentado que los **datos** tienen distintos tipos, pero ¿habría forma de verificar que **tipo de dato** es una **variable**?.<br>
Para esto podemos utilizar la función **type()** que devuelve el **tipo de dato** del **valor** o **variable** que se le pasa dentro de los paréntesis.

```python
type(2)
```

```python
# para este ejemplo cambiamos el valor de dos_str
dos_str = '2'
type(dos_str)
```
