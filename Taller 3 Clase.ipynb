- función **bool()**: devuelve en **bool** el **valor** o **contenido de  variable** que se le pasa dentro de los paréntesis.<br>


## Operadores aritméticos
Los **operadores aritméticos** son **caracteres especiales** que se utilizan para desempeñar operaciones matemáticas. Estas operaciones pueden ser realizadas entre variables de **tipo int o float**.
- **suma:** para la operación matemática de la suma se utiliza el **caracter especial** del más **(+)**.

```python
1+1
```

- **resta:** para la operación matemática de la resta se utiliza el **caracter especial** del menos o *guión* **(-)**.

```python
1-1
```

- **multiplicación:** para la operación matemática de la multiplicación se utiliza el **caracter especial** del asterisco **(*)**.

```python
2*2
```

- **división:** para la operación matemática de la división se utiliza el **caracter especial** del *slash* **(/)**.<br>

**TIP:** el **resultado** de la división convensional es un número **float** o *decimal* aunque el *resultado* sea entero.

```python
4/2
```

```python
3/2
```

```python
10/3
```

```python
3.6/2
```

- **división entera:** para que el **resultado** de la operación de división sea **int** o *entera* se utiliza el **caracter especial** de doble *slash* **(//)**.<br>

**TIP:** el **resultado** de la división entera me devuelve en **int** o *entero* la parte entera de la división *(no utiliza la propiedad de redondeo).*

```python
4//2
```

```python
3//2
```

```python
3.6//2
```

- **módulo:** la operación *módulo* nos devuelve el residuo de una división entre 2 números el **caracter especial** para esta operación es el porcentaje **(%)**.

```python
4%2
```

```python
3%2
```

```python
182%2
```

```python
1561%2
```

**TIP:** Las operaciones respetan las leyes matemáticas es decir si dividimos para **cero** Python devuelve un error.

```python
3//0
```

```python
4/0
```

```python
5%0
```

- **potencia:** para la operación matemática de la potencia se utiliza el **caracter especial** doble asterisco **(**** **)**.

```python
4**2
```

```python
4**3
```

```python
16**(1/2)
```

```python
64**(1/3)
```

**TIP:** El resultado de estas operaciones puede ser almacenado en variables o incluso ser incluidos dentro de los *paréntesis* de las funciones.

```python
x = 2 + 2
```

```python
y = 2 * 8
```

```python
print(3%2)
```

```python
4/2
```

```python
type(4/2)
```

**TIP:** Las operaciones aritméticas pueden ser realizadas entre **variables** e incluso combinado entre **variables y valores numéricos**

```python
print(y-x)
```

```python
print(y//4)
```

**TIP:** En una operación aritmética si alguno de los operandos no es numérico python devuelve un **error**

```python
print(3+"3")
```

```python
print(3+"tres")
```

**TIP:** pero si convertimos un tipo a numérico la operación podría realizarse.

```python
print(3 + int('3'))
```

### *Atajos*
Es muy común que en nuestro código reutilicemos una **varible** cambiando su valor a partir de **operaciones aritméticas**<br>
Por ejemplo: Supongamos que el euro tiene un **valor** de conversión al dólar de **1.2** para una primera semana. Pero a la semana siguiente dicho valor sube en **0.5**. ¿Cómo se representaría en python?

```python
dolar = 1
euro = 1.2

# Imprimir la tasa de conversión de la primera semana
print(dolar * euro)

```

```python
# La segunda semana incrementamos en 0.5 la tasa de conversión de la 1era semana
euro = euro + 0.5

# Imprimir la NUEVA tasa de conversión
print(dolar * euro)
```

- **atajo +=**.

```python
dolar = 1
euro = 1.2

# Imprimir la tasa de conversión de la primera semana
print(dolar * euro)

```

```python
# La segunda semana incrementamos en 0.5 la tasa de conversión de la 1era semana
# utilizar += es lo mismo que euro = euro + 0.5
euro += 0.5


# Imprimir la NUEVA tasa de conversión
print(dolar * euro)
```

Siguiendo el ejemplo del euro, supongamos que el euro ahora redujo su **valor** en **1** con respecto al dólar. ¿Cómo se representaría en python?

```python
euro = 1.2
euro -= 1

# print(euro)
print("El nuevo valor del euro es: ", euro)
```

```python
euro = 1.7
# 1 ocurre una operación en la parte derecha del igual (euro - 1)
# 2 se asigna el resultado a la variable del lado izquierdo
# euro = euro -1
euro -= 1
print(euro -= 1)
```

- **atajo -=**.

```python
#primera semana
dolar = 1
euro = 1.2
print(dolar * euro)

#segunda semana
euro += 0.5
print(dolar * euro)

#tercera semana
euro -= 1
print(dolar * euro)
```

```python
dolar = 1
euro = dolar
print(euro)
```

Ahora, supongamos que el euro duplico su **valor** con respecto al dólar. ¿Cómo se representaría en python?

- **atajo *=**.

```python
# primera_semana
dolar = 1
euro = 1.2
print(euro*dolar)

#segunda_semana
euro += 0.5
print(euro*dolar)

#tercera_semana
euro -= 1
print(euro*dolar)

# cuarta semana
euro *= 2
print(euro*dolar)
```

La siguiente semana euro vale una tercera parte lo que antes con respecto al dólar. ¿Cómo se representaría en python?

- **atajo /=**.

```python
euro = 1.4
euro /=3
print(euro)
```
