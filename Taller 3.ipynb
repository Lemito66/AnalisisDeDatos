
## Operadores aritméticos
Los **operadores aritméticos** son **caracteres especiales** que se utilizan para desempeñar operaciones matemáticas. Estas operaciones pueden ser realizadas entre variables de **tipo int o float**.
- **suma:** para la operación matemática de la suma se utiliza el **caracter especial** del más **(+)**.

```python
1+1
```

- **resta:** para la operación matemática de la resta se utiliza el **caracter especial** del menos o *guión* **(-)**.

```python
1-1
```

- **multiplicación:** para la operación matemática de la multiplicación se utiliza el **caracter especial** del asterisco **(*)**.

```python
2*2
```

- **división:** para la operación matemática de la división se utiliza el **caracter especial** del *slash* **(/)**.<br>
**TIP:** el **resultado** de la división convensional es un número **float** o *decimal* aunque el *resultado* sea entero.

```python
4/2
```

```python
3/2
```

- **división entera:** para que el **resultado** de la operación de división sea **int** o *entera* se utiliza el **caracter especial** de doble *slash* **(//)**.<br>
**TIP:** el **resultado** de la división entera me devuelve en **int** o *entero* la parte entera de la división *(no utiliza la propiedad de redondeo).*

```python
4//2
```

```python
3//2
```

- **módulo:** la operación *módulo* nos devuelve el residuo de una división entre 2 números el **caracter especial** para esta operación es el porcentaje **(%)**.

```python
4%2
```

```python
3%2
```

**TIP:** Las operaciones respetan las leyes matemáticas es decir si dividimos para **cero** Python devuelve un error.

```python
3//0
```

```python
4/0
```

```python
5%0
```

- **potencia:** para la operación matemática de la potencia se utiliza el **caracter especial** doble asterisco **(**** **)**.

```python
4**2
```

**TIP:** El resultado de estas operaciones puede ser almacenado en variables o incluso ser incluidos dentro de los *paréntesis* de las funciones.

```python
x = 2 + 2
```

```python
y = 2 * 8
```

```python
print(3%2)
```

```python
type(4/2)
```

**TIP:** Las operaciones aritméticas pueden ser realizadas entre **variables** e incluso combinado entre **variables y valores numéricos**

```python
print(y-x)
```

```python
print(y//4)
```

**TIP:** En una operación aritmética si alguno de los operandos no es numérico python devuelve un **error**

```python
print(3+"tres")
```

```python
print(3+"3")
```

**TIP:** pero si convertimos un tipo a numérico la operación podría realizarse.

```python
print(3 + int('3'))
```

## Operadores textuales
Los **operadores textuales** son **caracteres especiales** que se utilizan para realizar acciones entre variables de tipo **str** o *string*.
- **salida de escape:** para esta operación se utiliza el **caracter especial** del *backslash* **(\\)**.

```python
print('a veces usamos comillas para expresar ciertas cosas')
```

```python
print('a veces usamos comillas para expresar 'ciertas' cosas')
```

```python
print("a veces usamos comillas para expresar 'ciertas' cosas")
```

```python
print('a veces usamos comillas para expresar \'ciertas\' cosas')
```

Pero y si quiero *imprimir* como se utiliza un *backslash* junto a una comilla en Python. ¿cómo hacerlo? <br>
**TIP:** se utilizaría *la salida de escape* para el *backslash* **(\\\\)**

```python
print('se utiliza backslash \\\' como "salida de escape"')
```

Que tal si quiero imprimir un texto debajo del otro en una misma función **print()**. ¿cómo hacerlo?<br>
**TIP:** se utilizaría *la salida de escape* con **n** **(\\n)**

```python
print('Esto va arriba \ny esto abajo')
```

Y wue tal si quiero imprimir un texto debajo del otro e indentarlo en una misma función **print()**. ¿cómo hacerlo?<br>
**TIP:** se utilizaría *la salida de escape* con **n** y *la salida de escape* con **t** **(\\n\\t)**

```python
print('Esto va arriba \n\ty esto abajo indentado')
```

Pero con el *bucle* **for** podemos *repetir* una serie de *tareas* para cada **elemento** de un **conjunto de datos (dataset)**. ¿Cómo podemos hacerlo? Hay que usar la siguiente sintaxis:
 - **for** *each element* **in** *list* **:**

Traduciendo sería:
 - **Para** *cada elemento* **en** *la lista* **:**

Las palabras en **negritas** deben ser usados *literalmente*, mientras que en cursivas se puede cambiar dependiendo del uso que se quiera dar. Por ejemplo si tengo una **lista** llamada *a_list*, donde dice *la lista* debo poner esta **variable**.

```python
lista = [1,3,5]

for elemento in lista:
    print(elemento)
```

3 Detalles que quiero comentar respecto a la *sintaxis* y *semántica* de este *bucle*:

1. La variable *"elemento"* en el ejemplo anterior, no es necesario definirla previamente, porque se va a *inicializar* con el primer **elemento** de la **lista** y va a cambiar en cada **ciclo**. Así mismo, cuando el **for** haya acabado, la variable *"elemento"* va a tener el **valor** del **último elemento** de la **lista**.
2. Al finalizar la *línea de código* del **for** hay que poner los **dos puntos (:)** para que sea escrito correctamente el *bucle* **for** no poner estos **dos puntos (:)** me generaría un *error de sintaxis*.
3. La *indentación* es clave en la *sintaxis* del *bucle* **for** ya que define las *sentencias* que se ejecutarán en cada *iteración* del *bucle* También se conoce estas sentencias como el **cuerpo del bucle**.

```python
for fila in data_set:
    print(fila)
```
