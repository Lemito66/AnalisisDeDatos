## Operadores aritméticos
### *Atajos*

```python
# primera_semana
dolar = 1
euro = 1.2
print(euro * dolar)

#segunda_semana
# euro = euro + 0.5
euro += 0.5
print(euro * dolar)

#tercera_semana
# euro = euro - 1
euro -= 1
print(euro * dolar)

# cuarta semana
# euro = euro * 2
euro *= 2
print(euro * dolar)
```

La siguiente semana euro vale una tercera parte lo que antes con respecto al dólar. ¿Cómo se representaría en python?

- **atajo /=**.

```python
# Quinta semana
#euro = euro / 3
euro /= 3
print(euro)
```

Finalmente el euro se potenció cubicamente con respecto al dólar. ¿Cómo se representaría en python?

```python
# sexta semana
```

- **atajo **=**.

```python
# sexta semana
#euro = 0.4666666666666666
euro **= 3
print(euro)
```

## Operadores textuales
Los **operadores textuales** son **caracteres especiales** que se utilizan para realizar acciones entre variables de tipo **str** o *string*.
- **salida de escape:** para esta operación se utiliza el **caracter especial** del *backslash* **(\\)**.

```python
print('a veces usamos comillas para expresar ciertas cosas')
```

```python
print('a veces usamos comillas para expresar 'ciertas' cosas')
```

```python
print("a veces usamos comillas para expresar 'ciertas' cosas")
```

```python
print('a veces usamos comillas para expresar "ciertas" cosas')
```

```python
print('a veces usamos comillas para expresar \'ciertas\'  cosas')
```

```python
print('a veces usamos comillas para expresar \\\'ciertas\'  cosas')
```

Pero y si quiero *imprimir* como se utiliza un *backslash* junto a una comilla en Python. ¿cómo hacerlo? <br>
**TIP:** se utilizaría *la salida de escape* para el *backslash* **(\\\\)**

```python
print('se utiliza backslash \\\' como "salida de escape"')
```

Que tal si quiero imprimir un texto debajo del otro en una misma función **print()**. ¿cómo hacerlo?<br>
**TIP:** se utilizaría *la salida de escape* con **n** **(\\n)**

```python
print("arriba")
print("abajo")
```

```python
print("arriba abajo")
```

```python
print("arriba\nabajo")
```

```python
print('Esto va arriba \ny esto abajo')
```

Y que tal si quiero imprimir un texto debajo del otro e indentarlo en una misma función **print()**. ¿cómo hacerlo?<br>
**TIP:** se utilizaría *la salida de escape* con **n** y *la salida de escape* con **t** **(\\n\\t)**

```python
print('Esto va arriba \n\ty esto abajo indentado')
```

Hay muchos otras funcionalidades con *backslash* pero estas podrían ser las más útiles.
- **concatenación:** para esta operación se utiliza el **caracter especial** del más **(+)**. Concatenar consiste en juntar un texto al lado del otro.

```python
print("hola" + " mundo")
```

```python
print("hola" + "mundo")
```

```python
mundo = "mundo"
print("hola " + mundo)
```

## 5. Listas, Tablas y ciclos for.
### Listas.
Las **listas** son un conjunto ordenado de **datos.** Estos **datos** pueden ser del mismo **tipo de dato** pero no es necesario, podrían ser **datos mixtos**, lo más importante es entender la lista como un conjunto de **datos ordenado.**


```python
# Declaración de una lista vacía.
lista = []

print(lista)
```

```python
# Declaración de una lista del mismo tipo de datos, en este caso int
lista = [1,    3,     5]
print(lista)
```


![tabla](https://drive.google.com/uc?export=view&id=13dLcRV2hbEp1_5XTchHQjCn9qW_zzB3S)

Viendo el ejemplo de la **tabla** de arriba, podemos fijarnos que cada **fila** contiene **tipos de datos distintos** por **columnas**. De esta manera podemos crear una **lista** que representa la **fila** número cero con los **datos** de esa **fila.**

```python
# Declaración de una lista con tipos de datos mixtos
fila_1 = ['Facebook', 0.0, 'USD', 2974676, 3.5]
print(fila_1)
```

Podemos notar que la **numeración** de  **filas** de la **tabla** empiezan en **cero (0)**, esto no es casualidad, sino para acostumbrarnos a la nomenclatura de **orden**.

#### Nomenclatura de orden.
![tabla](https://drive.google.com/uc?export=view&id=1QO2gxFb-vxRDgsJqOEG-dr6JHVvxecvK)

Las **listas** y otros **tipos de datos grupales** en python utilizan **índices** para acceder a los distintos **elementos** de estos **tipos de datos.** En el caso de las **listas** que son **tipos de datos ordenados** estos **índices** mantienen una **nomenclatura de orden** que sería la siguiente:

![tabla](https://drive.google.com/uc?export=view&id=1bei9c8KMSHg2mR-H7svj7whCUg2OPkmL)

para acceder al **primer elemento de la lista** se lo hace a través del **índice cero (0)** y así en adelante, es decir, el **segundo elemento** se refiere con el **índice uno (1).** De esta manera, si mi **lista** tiene 5 **elementos** el último elemento se lo accedería con el **índice cuatro (4)**

![tabla](https://drive.google.com/uc?export=view&id=1Dk5WJFuCkQ2blckAUjHZvMfJwmoKQSxs)

```python
# mostrar el primer elemento de fila_1
# fila_1 = ['Facebook', 0.0, 'USD', 2974676, 3.5]
print(fila_1[0])
```

```python
# mostrar el segundo elemento de fila_1
print(fila_1[1])
```

```python
# mostrar el último elemento de fila_1
print(fila_1[4])
```

Al acceder a los **datos** de una lista, podemos tratarlos como **variables** por lo tanto podemos aplicar **operaciones** sobre estos **datos** como lo visto anteriormente, es decir **operaciones aritméticas o textuales** según convenga.

#### Longitud de listas.
Para conocer la **longitud** de una **lista** o de cualquier **tipo de dato** con **longitud** se utiliza la **función len()** poniendo entre los paréntesis el **dato** al cual queremos conocer su **longitud.**

![tabla](https://drive.google.com/uc?export=view&id=1abmyvtoSXPrOgkqWy_SWE08N0mkOCmRu)

```python
print(len(fila_1))
```

```python
lista_1 = [1,2,3]
print(len(lista_1))
```

```python
lista_2 = []
print(len(lista_2))
```

**TIP:** Conociendo la **longitud** de mi **lista** puedo llegar al **último elemento** de la misma.

```python
# fila_1[4]
tamano_fila_1 = len(fila_1)
ultimo_elemento = fila_1[tamano_fila_1]
print(ultimo_elemento)
```

```python
# fila_1[4]
tamano_fila_1 = len(fila_1)
#print(f"tamaño de fila_1 {tamano_fila_1}")
print("tamaño de fila_1 " + str(tamano_fila_1))
ultimo_elemento = fila_1[tamano_fila_1 - 1]
print(f"último elemento: {ultimo_elemento}")
```
