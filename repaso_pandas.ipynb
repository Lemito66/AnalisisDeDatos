Como saber cuantas marcas de carros existen

```python
def get_how_many_cars_exist(dataframe):
    dictionary_of_cars = {}
    for car in data[dataframe]:
        if car in dictionary_of_cars:
            dictionary_of_cars[car] += 1
        else:
            dictionary_of_cars[car] = 1
    return dictionary_of_cars
```

```python
get_how_many_cars_exist('car name')
```
